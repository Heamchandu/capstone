```python
df = train
```

```python
for i in df.columns:
  if df[i].nunique()<=7:
    df[i]=df[i].astype(str)
```

```python
df.info()
```

```python
for i in df.columns:
  if df[i].dtype == int:
    sns.distplot(df[i])
    plt.show()
```

```python
for i in df.columns:
  if df[i].dtype == int:
    sns.boxplot(x= df[i],y = df['Cover_Type'])
    plt.show()
  else:
    sns.countplot(x='Cover_Type', hue= i, data=df, dodge=True)
    plt.show()
```

```python
df_num =  df.select_dtypes(include = int)
df_num.head()
```

```python
sns.boxplot(df_num)
plt.xticks(rotation = 90)
plt.show()
```

```python
for i in df_num.columns:
  mean = df_num[i].mean()
  median =  df_num[i].median()
  print(i)
  print('mean: ', mean)
  print('median: ',median)
```
